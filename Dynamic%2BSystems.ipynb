```python
import numpy as np
import sys
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10,6)
```

Iteration: Simple Population Growth Model

n = population
n_0 = intital popultation
n_1 = population at year 1
n_t = pop at year t
birthrate (br) =  number of offspring produced each year
n_1 = br * n_0
n_2 = br * n_1

General model

n_t + 1 = br * n_t


```python
#initial settings
int_pop = 1
birth_rate = 2
time_step = 5
```

```python
def Population(int_pop, birth_rate, time_step):
    
    out = np.array([])
    out = np.append(out, int_pop)
    last_pop = np.array([])
    last_pop = np.append(last_pop, int_pop)  # track last years pop
    while time_step > 0:
        new_pop = birth_rate * int_pop
        out = np.append(out, [new_pop])
        last_pop = np.append(last_pop, [int_pop])
        int_pop = new_pop
        time_step = time_step - 1
    
    return out, last_pop
```

```python
#output
bunnie_count, last_years_bunnies = (Population(int_pop, birth_rate, time_step))
print (bunnie_count) 
print (last_years_bunnies)
```

This year population is Bunnie Count

```python
plt.plot(bunnie_count)
plt.title('Simple Population growth')
plt.ylabel('population')
plt.xlabel('Time')
plt.ylim(ymin=0)  
plt.xlim(xmin=0)
plt.show
```

```python
plt.title('Last Year Growth v This years')
plt.ylabel('This Year')
plt.xlabel('Last Year')

plt.plot(last_years_bunnies,bunnie_count)
plt.show()

print ('bunnie_count      ', bunnie_count) 
print ('last_years_bunnies',last_years_bunnies)
```
